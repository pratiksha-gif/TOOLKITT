**Q2)**In NumPy, broadcasting is a powerful mechanism that allows you to perform arithmetic operations on arrays of different shapes, without explicitly reshaping them. It makes operations more memory efficient and simplifies code.

The concept of broadcasting follows a few basic rules:

1. Align Shapes from the Right
To broadcast, NumPy compares the shapes of the arrays element-wise starting from the last dimension. If the dimensions are different, it tries to make them compatible according to the following rules.
2. Rules of Broadcasting
Broadcasting follows these steps:

Rule 1: If the arrays have a different number of dimensions, the smaller-dimensional array is padded with ones on the left side to match the number of dimensions of the larger array.

Rule 2: If the size of a dimension in one array is different from the size in the other array, one of the following must be true:

The size of the dimension in one of the arrays is 1 (in this case, NumPy will "stretch" it along that axis).
The size of the dimension is the same in both arrays.
Rule 3: The resulting shape of the broadcasted arrays is the element-wise maximum of the input shapes. If the shapes differ in length, the shorter shape is padded with ones.

**Q6)**1. Type of Data Storage
NumPy Arrays: Store elements of the same type (e.g., all integers, all floats) in a contiguous block of memory, which makes them more efficient for numerical computations.
Python Lists: Can store elements of different types (e.g., integers, strings, and other objects) and are implemented as arrays of pointers to objects.
2. Performance
NumPy Arrays: Highly optimized for performance. Operations on NumPy arrays are faster than on Python lists because they are implemented in C, and operations are vectorized.
Python Lists: Slower for mathematical or numerical operations because of their flexibility in holding mixed data types and the overhead of dynamic resizing.
3. Memory Efficiency
NumPy Arrays: More memory-efficient as they use a contiguous block of memory to store data, with less overhead per element.
Python Lists: Less memory-efficient because each element in a list is a pointer to a Python object, which adds overhead.
4. Operations (Vectorization)
NumPy Arrays: Support element-wise operations (e.g., addition, multiplication) directly without the need for explicit loops. NumPy performs these operations using optimized C code.
Python Lists: Do not support direct element-wise operations. You would need to loop through the list to perform an operation on each element.
5. Functionality
NumPy Arrays: Offer a wide range of mathematical, statistical, and linear algebra functions built-in (e.g., matrix operations, FFT, broadcasting).
Python Lists: Offer basic operations like appending, slicing, and iteration, but lack advanced mathematical functions.
6. Dimensionality
NumPy Arrays: Can be multi-dimensional (e.g., 2D, 3D arrays) and support advanced features like reshaping, transposing, and broadcasting.
Python Lists: Essentially one-dimensional, though lists of lists can be used to simulate multi-dimensional structures (though this is inefficient compared to NumPy arrays).
7. Syntax
NumPy Arrays: You can use powerful slicing and indexing features. For example, NumPy supports slicing in multiple dimensions.
Python Lists: Have basic slicing capabilities but are limited in handling multi-dimensional data in an efficient way.
8. Support for Null Values
NumPy Arrays: Use numpy.nan for missing values in arrays of floats, and can handle missing data more effectively.
Python Lists: Can have None values to represent missing data but dont have built-in support for numerical missing values.
9. Size Flexibility
NumPy Arrays: Once created, their size is fixed. To change the size, a new array must be created.
Python Lists: Can dynamically grow or shrink as needed (via append() or remove() operations).

**Q17)**
In Pandas, both apply() and map() are used to apply a function to data, but they are used in different contexts and have distinct differences. Here's a breakdown:

1. apply()
Scope: Works on DataFrames or Series.
Functionality: Applies a function along an axis (rows or columns for DataFrames) or to each element in a Series.
Flexibility: More flexible, as it can apply functions that operate on rows, columns, or the entire Series/DataFrame.
Input Function: Can accept functions that return single values or multiple values.

2. map()
Scope: Works only on Series.
Functionality: Maps a function, dictionary, or Series to transform each element in the Series.
Flexibility: Limited compared to apply(), as it works element-wise only.
Input: Accepts functions, dictionaries for value mapping, or another Series for alignment-based transformation.
